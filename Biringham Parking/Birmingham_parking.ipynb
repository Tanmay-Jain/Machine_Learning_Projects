```python
# creating new dataframe with only carapark info 
carpark = pd.DataFrame(data.groupby(by=['SystemCodeNumber'])['LastUpdated'].min())
```

```python
carpark.columns = ['Start_Date']
```

```python
carpark.head()
```

```python
carpark['End_Date'] = data.groupby(by=['SystemCodeNumber'])['LastUpdated'].max()
```

```python
carpark.head()
```

```python
carpark['Max_Occupancy'] = data.groupby(by=['SystemCodeNumber'])['Occupancy'].max()
```

```python
carpark.head()
```

```python
carpark['Min_Occupancy'] = data.groupby(by=['SystemCodeNumber'])['Occupancy'].min()
```

```python
carpark.head()
```

```python
carpark['Avg_Occupancy'] = data.groupby(by=['SystemCodeNumber'])['Occupancy'].mean()
```

```python
carpark.head()
```

```python
carpark['Capacity'] = data.groupby(by=['SystemCodeNumber'])['Capacity'].unique()
```
